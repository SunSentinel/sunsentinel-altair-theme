# Sun Sentinel Altair theme

This notebook is meant to show how to use this theme for the Altair data visualization library based on the graphics styles and brand guidelines at the South Florida Sun Sentinel.

Also included in this notebook are examples of Altair charts and visualizations using this theme. We used [Altair's example gallery](https://altair-viz.github.io/gallery/index.html) for this.

This theme was inspired by [husky_theme](https://github.com/deppen8/husky-altair-theme).

#### Import packages for this example

```python
import altair as alt
from vega_datasets import data
from altair.expr import datum, if_
alt.renderers.enable('notebook')

import pandas as pd
import numpy as np
```

#### Enable the theme

```python
# Register the custom theme and give it a name.
alt.themes.register('sunsentinel', sunsentinel_theme)

# Enable the theme
alt.themes.enable('sunsentinel')
```

## Examples

### Scatter plot

```python
cars = data.cars()

alt.Chart(cars).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Origin', 'Horsepower', 'Miles_per_Gallon']
).properties(
    title="Simple Scatter Plot with Legend"
)
```

### Simple bar chart

```python
simple_bar_data = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': [28, 55, 43, 91, 81, 53, 19, 87, 52]
})

alt.Chart(simple_bar_data).mark_bar().encode(
    x='a',
    y='b'
)
```

### Grouped bar chart

```python
groupedbar_data = data.population.url

alt.Chart(groupedbar_data).mark_bar(stroke='transparent').encode(
    alt.X('gender:N', scale=alt.Scale(rangeStep=12), axis=alt.Axis(title='')),
    alt.Y('sum(people):Q', axis=alt.Axis(title='population', grid=False)),
    color=alt.Color('gender:N'),
    column='age:O'
).configure_view(
    stroke='transparent'
).configure_axis(
    domainWidth=0.8
).transform_filter(
    datum.year == 2000
).transform_calculate(
    'gender', if_(datum.sex == 2, 'Female', 'Male')
)
```

### Stacked Bar Chart

```python
barley = data.barley()

alt.Chart(barley).mark_bar().encode(
    x='variety',
    y='sum(yield)',
    color='site'
)
```
